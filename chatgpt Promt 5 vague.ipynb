```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the dataset
superstore_data = pd.read_csv('Superstore.csv', encoding='latin1')

# Set up the path to save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "P4V1"

# Create the folder if it does not exist
save_path.mkdir(parents=True, exist_ok=True)

# Group by Customer Segment and calculate total sales and profit
segment_data = superstore_data.groupby('Segment').agg(
    total_sales=('Sales', 'sum'),
    total_profit=('Profit', 'sum')
).reset_index()

# Plot the data
fig, ax = plt.subplots(1, 2, figsize=(14, 7))

# Plot Total Sales by Segment
ax[0].bar(segment_data['Segment'], segment_data['total_sales'], color='blue')
ax[0].set_title('Total Sales by Customer Segment')
ax[0].set_xlabel('Customer Segment')
ax[0].set_ylabel('Total Sales')
ax[0].tick_params(axis='x', rotation=45)

# Plot Total Profit by Segment
ax[1].bar(segment_data['Segment'], segment_data['total_profit'], color='green')
ax[1].set_title('Total Profit by Customer Segment')
ax[1].set_xlabel('Customer Segment')
ax[1].set_ylabel('Total Profit')
ax[1].tick_params(axis='x', rotation=45)

# Tight layout to prevent overlap
plt.tight_layout()

# Save the figure to the specified location
plot_filename = save_path / 'trial1.png'
plt.savefig(plot_filename)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load dataset
superstore_data = pd.read_csv('Superstore.csv', encoding='latin1')

# Group by Segment and sum sales and profit
segment_summary = superstore_data.groupby('Segment')[['Sales', 'Profit']].sum().reset_index()

# Plot setup
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar width and positions
bar_width = 0.35
index = range(len(segment_summary))

# Bars for Sales and Profit
bar1 = ax1.bar(index, segment_summary['Sales'], bar_width, label='Sales', color='skyblue')
bar2 = ax1.bar([i + bar_width for i in index], segment_summary['Profit'], bar_width, label='Profit', color='salmon')

# Labels and title
ax1.set_xlabel('Customer Segment')
ax1.set_ylabel('Amount in USD')
ax1.set_title('Comparison of Sales and Profit by Customer Segment')
ax1.set_xticks([i + bar_width / 2 for i in index])
ax1.set_xticklabels(segment_summary['Segment'])
ax1.legend()

# Annotate bars with values
for bars in [bar1, bar2]:
    for bar in bars:
        height = bar.get_height()
        ax1.annotate(f'{height:,.0f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=8)

# Set up the path to save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "P5V1"
save_path.mkdir(parents=True, exist_ok=True)

# Save the plot
plot_filename = save_path / "trial4.png"
plt.tight_layout()
plt.savefig(plot_filename)
plt.show()
```
