```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the dataset
superstore_data = pd.read_csv('Superstore.csv', encoding='latin1')

# Group by 'Segment' and calculate average sales and profit
grouped_data = superstore_data.groupby('Segment')[['Sales', 'Profit']].mean()

# Plotting the grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar chart
grouped_data.plot(kind='bar', ax=ax)

# Set the title and labels
ax.set_title('Average Sales and Profit by Customer Segment', fontsize=16)
ax.set_xlabel('Customer Segment', fontsize=12)
ax.set_ylabel('Amount', fontsize=12)
ax.set_xticklabels(grouped_data.index, rotation=0)

# Set up the path to save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "P5V2"

# Create the folder if it doesn't exist
save_path.mkdir(parents=True, exist_ok=True)

# Save the plot as a PNG file
file_path = save_path / "trial1.png"
plt.tight_layout()
plt.savefig(file_path)

# Show the plot
plt.show()

print(f"Plot saved as {file_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the dataset
superstore_data = pd.read_csv('Superstore.csv', encoding='latin1')

# Calculate the average sales and profit by segment
segment_avg = superstore_data.groupby('Segment')[['Sales', 'Profit']].mean().reset_index()

# Create a grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.35  # Bar width

# Plot the average sales and profit as grouped bars
ax.bar(segment_avg['Segment'] - width/2, segment_avg['Sales'], width, label='Average Sales', color='blue')
ax.bar(segment_avg['Segment'] + width/2, segment_avg['Profit'], width, label='Average Profit', color='green')

# Add labels and title
ax.set_xlabel('Customer Segment')
ax.set_ylabel('Amount ($)')
ax.set_title('Average Sales and Profit by Customer Segment')
ax.set_xticks(range(len(segment_avg['Segment'])))
ax.set_xticklabels(segment_avg['Segment'])
ax.legend()

# Set up the path to save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "P5V2"
save_path.mkdir(parents=True, exist_ok=True)  # Create folder if it doesn't exist

# Save the plot as a file
file_name = save_path / "trial2.png"
plt.tight_layout()  # Adjust the layout to prevent clipping
plt.savefig(file_name)

# Display the plot
plt.show()

print(f"Graph saved as {file_name}")
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the dataset
superstore_data = pd.read_csv('Superstore.csv', encoding='latin1')

# Calculate average sales and profit by customer segment
avg_sales_profit = superstore_data.groupby('Segment').agg({'Sales': 'mean', 'Profit': 'mean'}).reset_index()

# Create the grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Set bar positions and width
bar_width = 0.35
index = range(len(avg_sales_profit))

# Create bars for Sales and Profit
bar1 = ax.bar(index, avg_sales_profit['Sales'], bar_width, label='Average Sales', color='blue')
bar2 = ax.bar([i + bar_width for i in index], avg_sales_profit['Profit'], bar_width, label='Average Profit', color='green')

# Labeling
ax.set_xlabel('Customer Segment')
ax.set_ylabel('Amount')
ax.set_title('Average Sales and Profit by Customer Segment')
ax.set_xticks([i + bar_width / 2 for i in index])  # position the ticks in the middle of the grouped bars
ax.set_xticklabels(avg_sales_profit['Segment'])
ax.legend()

# Set up the path to save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "P5V2"
save_path.mkdir(parents=True, exist_ok=True)  # Create folder if it doesn't exist

# Save the plot to the specified folder
save_file = save_path / "trial6.png"
plt.savefig(save_file)

# Show the plot
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the dataset (ensure the file path is correct for your system)
superstore_data = pd.read_csv('Superstore.csv', encoding='latin1')

# Calculate average sales and profit by customer segment
avg_sales_profit = superstore_data.groupby('Segment').agg(
    avg_sales=('Sales', 'mean'),
    avg_profit=('Profit', 'mean')
).reset_index()

# Set up the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Create a grouped bar chart
bar_width = 0.35
index = range(len(avg_sales_profit))

# Plot average sales and average profit
bar1 = ax.bar(index, avg_sales_profit['avg_sales'], bar_width, label='Average Sales', color='b')
bar2 = ax.bar([i + bar_width for i in index], avg_sales_profit['avg_profit'], bar_width, label='Average Profit', color='g')

# Set the labels and title
ax.set_xlabel('Customer Segment')
ax.set_ylabel('Amount')
ax.set_title('Average Sales and Profit by Customer Segment')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(avg_sales_profit['Segment'])
ax.legend()

# Set up the path to save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "P5V2"

# Create the directory if it doesn't exist
save_path.mkdir(parents=True, exist_ok=True)

# Save the plot as 'trial8.png' in the P5V2 folder
plot_filename = save_path / "trial8.png"
plt.tight_layout()
plt.savefig(plot_filename)

# Show the plot
plt.show()

print(f"Plot saved as {plot_filename}")
```
