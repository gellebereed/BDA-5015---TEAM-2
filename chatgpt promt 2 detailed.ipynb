```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the data
superstore_data = pd.read_csv('Superstore.csv', encoding='latin1')

# Aggregate total profit by region
profit_by_region = superstore_data.groupby('Region')['Profit'].sum().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(8, 6))
profit_by_region.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.tight_layout()

# Save the plot to the specified path
desktop = Path.home() / "Desktop"
save_path = desktop / "P2V2"
save_path.mkdir(parents=True, exist_ok=True)  # Ensure the folder exists

plt.savefig(save_path / "Trial1.png")
plt.show()
```
