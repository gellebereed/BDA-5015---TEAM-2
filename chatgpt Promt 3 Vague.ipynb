```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the data (adjust the file path as necessary)
superstore_data = pd.read_csv('Superstore.csv', encoding='latin1')

# Ensure that 'Discount' and 'Profit' are in the correct numeric format
superstore_data['Discount'] = pd.to_numeric(superstore_data['Discount'], errors='coerce')
superstore_data['Profit'] = pd.to_numeric(superstore_data['Profit'], errors='coerce')

# Create a scatter plot to explore the relationship between Discount and Profit
plt.figure(figsize=(10, 6))
plt.scatter(superstore_data['Discount'], superstore_data['Profit'], alpha=0.5, color='blue')

# Set plot titles and labels
plt.title('Relationship between Discount and Profit', fontsize=16)
plt.xlabel('Discount', fontsize=12)
plt.ylabel('Profit', fontsize=12)

# Add a line of best fit (optional) for better understanding of the trend
# Fit a linear regression model using numpy for the line of best fit
import numpy as np
slope, intercept = np.polyfit(superstore_data['Discount'].dropna(), superstore_data['Profit'].dropna(), 1)
plt.plot(superstore_data['Discount'], slope * superstore_data['Discount'] + intercept, color='red', lw=2, label='Fit Line')

# Save the figure to the specified path on your Desktop
desktop = Path.home() / "Desktop"
save_path = desktop / "P3V1"

# Create the directory if it doesn't exist
save_path.mkdir(parents=True, exist_ok=True)

# Save the plot
plt.legend()
plt.savefig(save_path / 'Trial4.png', bbox_inches='tight')

# Show the plot
plt.show()
```
