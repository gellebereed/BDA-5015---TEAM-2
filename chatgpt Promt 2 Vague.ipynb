```python
from pathlib import Path
import os
import pandas as pd
import matplotlib.pyplot as plt

# Masaüstündeki "AAAAA" klasörünü oluştur
desktop = Path.home() / "Desktop"
save_path = desktop / "AAAAA"
os.makedirs(save_path, exist_ok=True)
plot_id = 1

def save_and_show():
    global plot_id
    plt.savefig(save_path / f'plot_{plot_id}.png')
    plot_id += 1
    plt.show()

# Veri yükle
superstore_data = pd.read_csv('Superstore.csv', encoding='latin1')

# Her bir grafik aşağıdaki gibi tanımlı (1–10 arası)

# Grafik 1
region_profit = superstore_data.groupby('Region')['Profit'].sum().reset_index()
region_profit_sorted = region_profit.sort_values('Profit', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(region_profit_sorted['Region'], region_profit_sorted['Profit'], color='skyblue')
plt.title('Total Profit Across Different Regions')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
save_and_show()

# Grafik 2
region_profit = superstore_data.groupby('Region')['Profit'].sum().reset_index()
region_profit_sorted = region_profit.sort_values(by='Profit', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(region_profit_sorted['Region'], region_profit_sorted['Profit'], color='skyblue')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.title('Total Profit Across Different Regions')
plt.xticks(rotation=45)
plt.tight_layout()
save_and_show()

# Grafik 3
region_profit = superstore_data.groupby('Region')['Profit'].sum().reset_index()
region_profit_sorted = region_profit.sort_values(by='Profit', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(region_profit_sorted['Region'], region_profit_sorted['Profit'], color='skyblue')
plt.title('Total Profit Across Different Regions', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Profit ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
save_and_show()

# Grafik 4
profit_by_region = superstore_data.groupby('Region')['Profit'].sum().reset_index()
profit_by_region_sorted = profit_by_region.sort_values(by='Profit', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(profit_by_region_sorted['Region'], profit_by_region_sorted['Profit'], color='skyblue')
plt.title('Total Profit by Region', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Profit ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
save_and_show()

# Grafik 5
region_profit = superstore_data.groupby('Region')['Profit'].sum().reset_index()
region_profit_sorted = region_profit.sort_values(by='Profit', ascending=False)
plt.figure(figsize=(10,6))
plt.bar(region_profit_sorted['Region'], region_profit_sorted['Profit'], color='skyblue')
plt.title('Total Profit by Region', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Profit ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
save_and_show()

# Grafik 6
profit_by_region = superstore_data.groupby('Region')['Profit'].sum().reset_index()
profit_by_region_sorted = profit_by_region.sort_values(by='Profit', ascending=False)
plt.figure(figsize=(10,6))
plt.bar(profit_by_region_sorted['Region'], profit_by_region_sorted['Profit'], color='skyblue')
plt.title('Total Profit by Region', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Profit ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
save_and_show()

# Grafik 7
region_profit = superstore_data.groupby('Region')['Profit'].sum().reset_index()
region_profit_sorted = region_profit.sort_values(by='Profit', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(region_profit_sorted['Region'], region_profit_sorted['Profit'], color='skyblue')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.title('Total Profit by Region')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
save_and_show()

# Grafik 8
profit_by_region = superstore_data.groupby('Region')['Profit'].sum().reset_index()
profit_by_region_sorted = profit_by_region.sort_values(by='Profit', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(profit_by_region_sorted['Region'], profit_by_region_sorted['Profit'], color='skyblue')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.title('Total Profit Across Different Regions')
plt.xticks(rotation=45)
plt.tight_layout()
save_and_show()

# Grafik 9
region_profit = superstore_data.groupby('Region')['Profit'].sum().reset_index()
region_profit_sorted = region_profit.sort_values(by='Profit', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(region_profit_sorted['Region'], region_profit_sorted['Profit'], color='skyblue')
plt.title('Profit Across Different Regions', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Profit ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
save_and_show()

# Grafik 10
profit_by_region = superstore_data.groupby('Region')['Profit'].sum().reset_index()
profit_by_region = profit_by_region.sort_values(by='Profit', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(profit_by_region['Region'], profit_by_region['Profit'], color='skyblue')
plt.title('Total Profit by Region', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Profit ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
save_and_show()
```
