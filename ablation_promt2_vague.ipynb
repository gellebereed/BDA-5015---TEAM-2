```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the raw dataset
superstore_data = pd.read_csv('Superstore_RAW.csv', encoding='latin1')

# Clean the dataset
superstore_data.columns = superstore_data.columns.str.strip().str.lower().str.replace(' ', '_')
superstore_data['order_date'] = pd.to_datetime(superstore_data['order_date'], errors='coerce')
superstore_data['profit'] = pd.to_numeric(superstore_data['profit'], errors='coerce')
superstore_data['region'] = superstore_data['region'].str.strip()

# Drop rows with missing critical data
superstore_data.dropna(subset=['region', 'profit'], inplace=True)

# Aggregate profit by region
region_profit = superstore_data.groupby('region')['profit'].sum().sort_values()

# Plotting
plt.figure(figsize=(10, 6))
region_profit.plot(kind='barh', color='skyblue')
plt.title('Total Profit by Region')
plt.xlabel('Total Profit')
plt.ylabel('Region')
plt.grid(axis='x')

# Save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "RAWP2V1"
save_path.mkdir(parents=True, exist_ok=True)
save_file_path = save_path / "Trial1.png"
plt.tight_layout()
plt.savefig(save_file_path)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the data
superstore_data = pd.read_csv('Superstore_RAW.csv', encoding='latin1')

# Clean the dataset
# - Standardize column names
superstore_data.columns = superstore_data.columns.str.strip().str.lower().str.replace(' ', '_')

# - Convert 'profit' to numeric
superstore_data['profit'] = pd.to_numeric(superstore_data['profit'], errors='coerce')

# - Clean region names and drop rows with missing values
superstore_data['region'] = superstore_data['region'].astype(str).str.strip()
superstore_data = superstore_data.dropna(subset=['profit', 'region'])

# Group by region and calculate total profit
region_profit = superstore_data.groupby('region')['profit'].sum().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
region_profit.plot(kind='bar')
plt.title('Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.grid(axis='y')

# Save the plot to Desktop
desktop = Path.home() / "Desktop"
save_path = desktop / "RAWP2V1"
save_path.mkdir(parents=True, exist_ok=True)

save_file_path = save_path / "Trial2.png"
plt.tight_layout()  # Adjust layout to prevent overlap
plt.savefig(save_file_path)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the data
superstore_data = pd.read_csv('Superstore_RAW.csv', encoding='latin1')

# Clean the dataset
# Standardize column names: lowercase, strip whitespace, replace spaces with underscores
superstore_data.columns = superstore_data.columns.str.strip().str.lower().str.replace(' ', '_')

# Drop rows with missing values in essential columns
essential_columns = ['region', 'profit']
superstore_data = superstore_data.dropna(subset=essential_columns)

# Convert profit to numeric
superstore_data['profit'] = pd.to_numeric(superstore_data['profit'], errors='coerce')
superstore_data = superstore_data.dropna(subset=['profit'])

# Group by region and sum profit
region_profit = superstore_data.groupby('region')['profit'].sum().sort_values(ascending=False)

# Plot the result
plt.figure(figsize=(10, 6))
region_profit.plot(kind='bar')
plt.title('Total Profit by Region')
plt.ylabel('Profit')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.tight_layout()

# Set up the file path to save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "RAWP2V1"
save_path.mkdir(parents=True, exist_ok=True)

# Save the plot as a PNG file
save_file_path = save_path / "Trial4.png"
plt.savefig(save_file_path)

# Show the plot
plt.show()
```
