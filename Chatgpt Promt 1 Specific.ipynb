```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Klasör oluşturma ve sayaç
save_path = '/Users/mustafakeremsani/Desktop/plots'
os.makedirs(save_path, exist_ok=True)
plot_id = 1

def save_and_show():
    global plot_id
    plt.savefig(f'{save_path}/plot_{plot_id}.png')
    plot_id += 1
    plt.show()

# Ortak veri hazırlığı
superstore_data = pd.read_csv('Superstore.csv', encoding='latin1')
superstore_data['Order Date'] = pd.to_datetime(superstore_data['Order Date'])

# Kod 1
mask = (superstore_data['Order Date'] >= '2016-01-01') & (superstore_data['Order Date'] <= '2018-12-31')
filtered_data = superstore_data.loc[mask]
filtered_data['YearMonth'] = filtered_data['Order Date'].dt.to_period('M')
monthly_sales = filtered_data.groupby('YearMonth')['Sales'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Sales'], marker='o', linestyle='-')
plt.title('Monthly Total Sales (2016 - 2018)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
save_and_show()

# Kod 2
filtered_data = superstore_data[
    (superstore_data['Order Date'].dt.year >= 2016) &
    (superstore_data['Order Date'].dt.year <= 2018)
]
filtered_data['Year-Month'] = filtered_data['Order Date'].dt.to_period('M')
monthly_sales = filtered_data.groupby('Year-Month')['Sales'].sum().reset_index()
monthly_sales['Year-Month'] = monthly_sales['Year-Month'].dt.to_timestamp()
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Year-Month'], monthly_sales['Sales'], marker='o', linestyle='-')
plt.title('Monthly Total Sales (2016–2018)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
save_and_show()

# Kod 3
mask = (superstore_data['Order Date'].dt.year >= 2016) & (superstore_data['Order Date'].dt.year <= 2018)
filtered_data = superstore_data[mask]
filtered_data['YearMonth'] = filtered_data['Order Date'].dt.to_period('M')
monthly_sales = filtered_data.groupby('YearMonth')['Sales'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Sales'], marker='o')
plt.title('Monthly Total Sales (2016 - 2018)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
save_and_show()

# Kod 4
mask = (superstore_data['Order Date'] >= '2016-01-01') & (superstore_data['Order Date'] <= '2018-12-31')
filtered_data = superstore_data.loc[mask]
filtered_data['Year-Month'] = filtered_data['Order Date'].dt.to_period('M').astype(str)
monthly_sales = filtered_data.groupby('Year-Month')['Sales'].sum().reset_index()
monthly_sales['Year-Month'] = pd.to_datetime(monthly_sales['Year-Month'])
monthly_sales.sort_values('Year-Month', inplace=True)
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Year-Month'], monthly_sales['Sales'], marker='o', linestyle='-')
plt.title('Monthly Total Sales (2016–2018)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
save_and_show()

# Kod 5
filtered_data = superstore_data.loc[
    (superstore_data['Order Date'].dt.year >= 2016) &
    (superstore_data['Order Date'].dt.year <= 2018)
]
filtered_data['YearMonth'] = filtered_data['Order Date'].dt.to_period('M')
monthly_sales = filtered_data.groupby('YearMonth')['Sales'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Sales'], marker='o', linestyle='-')
plt.title('Monthly Total Sales (2016–2018)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
save_and_show()

# Kod 6
filtered_data = superstore_data[
    (superstore_data['Order Date'].dt.year >= 2016) &
    (superstore_data['Order Date'].dt.year <= 2018)
]
filtered_data['Year-Month'] = filtered_data['Order Date'].dt.to_period('M')
monthly_sales = filtered_data.groupby('Year-Month')['Sales'].sum().reset_index()
monthly_sales['Year-Month'] = monthly_sales['Year-Month'].astype(str)
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Year-Month'], monthly_sales['Sales'], marker='o', linestyle='-')
plt.title('Monthly Total Sales (2016–2018)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
save_and_show()

# Kod 7
mask = (superstore_data['Order Date'].dt.year >= 2016) & (superstore_data['Order Date'].dt.year <= 2018)
filtered_data = superstore_data[mask]
filtered_data['Year-Month'] = filtered_data['Order Date'].dt.to_period('M')
monthly_sales = filtered_data.groupby('Year-Month')['Sales'].sum().reset_index()
monthly_sales['Year-Month'] = monthly_sales['Year-Month'].dt.to_timestamp()
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Year-Month'], monthly_sales['Sales'], marker='o', linestyle='-')
plt.title('Monthly Total Sales (2016 - 2018)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
save_and_show()

# Kod 8
filtered_data = superstore_data[
    (superstore_data['Order Date'].dt.year >= 2016) & 
    (superstore_data['Order Date'].dt.year <= 2018)
]
filtered_data['YearMonth'] = filtered_data['Order Date'].dt.to_period('M')
monthly_sales = filtered_data.groupby('YearMonth')['Sales'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Sales'], marker='o', linestyle='-')
plt.title('Monthly Total Sales (2016–2018)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
save_and_show()

# Kod 9
filtered_data = superstore_data[
    (superstore_data['Order Date'].dt.year >= 2016) &
    (superstore_data['Order Date'].dt.year <= 2018)
]
filtered_data['YearMonth'] = filtered_data['Order Date'].dt.to_period('M')
monthly_sales = filtered_data.groupby('YearMonth')['Sales'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Sales'], marker='o', linestyle='-')
plt.title('Monthly Total Sales (2016 - 2018)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
save_and_show()

# Kod 10
filtered_data = superstore_data[
    (superstore_data['Order Date'].dt.year >= 2016) &
    (superstore_data['Order Date'].dt.year <= 2018)
]
filtered_data['YearMonth'] = filtered_data['Order Date'].dt.to_period('M')
monthly_sales = filtered_data.groupby('YearMonth')['Sales'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Sales'], marker='o', linestyle='-')
plt.title('Monthly Total Sales (2016 - 2018)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
save_and_show()
```
