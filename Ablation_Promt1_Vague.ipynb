```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the data
superstore_data = pd.read_csv('Superstore_RAW.csv', encoding='latin1')

# Clean the dataset
# Rename columns: lowercase, strip whitespace, replace spaces with underscores
superstore_data.columns = superstore_data.columns.str.strip().str.lower().str.replace(' ', '_')

# Convert 'order_date' to datetime format
superstore_data['order_date'] = pd.to_datetime(superstore_data['order_date'], errors='coerce')

# Drop rows with missing order_date or sales
superstore_data = superstore_data.dropna(subset=['order_date', 'sales'])

# Convert sales column to numeric
superstore_data['sales'] = pd.to_numeric(superstore_data['sales'], errors='coerce')
superstore_data = superstore_data.dropna(subset=['sales'])

# Set the order_date as the index for resampling
superstore_data.set_index('order_date', inplace=True)

# Resample sales data monthly and sum sales
monthly_sales = superstore_data['sales'].resample('M').sum()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', linewidth=2)
plt.title('Monthly Sales Trend Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Sales (USD)', fontsize=12)
plt.grid(True)

# Set up the file path to save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "RAWP1V1"
save_path.mkdir(parents=True, exist_ok=True)

# Save the plot as a PNG file
save_file_path = save_path / "Trial1.png"
plt.tight_layout()  # Adjust layout to prevent overlap
plt.savefig(save_file_path)

# Show the plot
plt.show()
```
