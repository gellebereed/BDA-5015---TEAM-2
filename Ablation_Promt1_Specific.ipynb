```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the data
superstore_data = pd.read_csv('Superstore_RAW.csv', encoding='latin1')

# Clean the dataset
# Normalize column names
superstore_data.columns = superstore_data.columns.str.strip().str.lower().str.replace(' ', '_')

# Convert 'order_date' to datetime
superstore_data['order_date'] = pd.to_datetime(superstore_data['order_date'], errors='coerce')

# Drop rows with missing order_date or sales
superstore_data = superstore_data.dropna(subset=['order_date', 'sales'])

# Convert sales column to numeric
superstore_data['sales'] = pd.to_numeric(superstore_data['sales'], errors='coerce')
superstore_data = superstore_data.dropna(subset=['sales'])

# Filter data between 2016 and 2018
filtered_data = superstore_data[
    (superstore_data['order_date'] >= '2016-01-01') & 
    (superstore_data['order_date'] <= '2018-12-31')
]

# Set order_date as index
filtered_data.set_index('order_date', inplace=True)

# Resample monthly and sum sales
monthly_sales = filtered_data['sales'].resample('M').sum()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title('Monthly Total Sales from 2016 to 2018')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)

# Set up the file path to save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "RAWP1V2"
save_path.mkdir(parents=True, exist_ok=True)

# Save the plot
save_file_path = save_path / "Trial1.png"
plt.tight_layout()
plt.savefig(save_file_path)

# Show the plot
plt.show()
```
