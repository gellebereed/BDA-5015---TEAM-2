```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the data
superstore_data = pd.read_csv('Superstore_RAW.csv', encoding='latin1')

# Convert 'SALES' to numeric and handle malformed entries
superstore_data['SALES'] = pd.to_numeric(superstore_data['SALES'], errors='coerce')

# Drop rows with missing values in critical columns
cleaned_data = superstore_data.dropna(subset=['SALES', 'DISCOUNT', 'profit'])

# Extract necessary columns
analysis_data = cleaned_data[['DISCOUNT', 'profit']]

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(analysis_data['DISCOUNT'], analysis_data['profit'], alpha=0.5)
plt.title('Impact of Discount on Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid(True)

# Set up the file path to save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "RAWP3V1"
save_path.mkdir(parents=True, exist_ok=True)
save_file_path = save_path / "Trial1.png"

# Save and show the plot
plt.tight_layout()
plt.savefig(save_file_path)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Veriyi yükle
superstore_data = pd.read_csv('Superstore_RAW.csv', encoding='latin1')

# Veriyi temizle: SALES sütununu sayısal yap ve eksik verileri kaldır
superstore_data['SALES'] = pd.to_numeric(superstore_data['SALES'], errors='coerce')
cleaned_data = superstore_data.dropna(subset=['SALES', 'profit'])

# İndirime göre ortalama satış ve kârı hesapla
discount_analysis = cleaned_data.groupby('DISCOUNT')[['SALES', 'profit']].mean().reset_index()

# Görselleştirme
plt.figure(figsize=(10, 6))
plt.plot(discount_analysis['DISCOUNT'], discount_analysis['SALES'], marker='o', label='Average Sales')
plt.plot(discount_analysis['DISCOUNT'], discount_analysis['profit'], marker='s', label='Average Profit')
plt.xlabel('Discount')
plt.ylabel('Amount')
plt.title('Effect of Discounts on Business Performance')
plt.legend()
plt.grid(True)

# Grafik kaydetme yolu
desktop = Path.home() / "Desktop"
save_path = desktop / "RAWP3V1"
save_path.mkdir(parents=True, exist_ok=True)
save_file_path = save_path / "Trial2.png"

# Grafiği kaydet ve göster
plt.tight_layout()
plt.savefig(save_file_path)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the data
superstore_data = pd.read_csv('Superstore_RAW.csv', encoding='latin1')

# Remove rows where 'SALES' is 'unknown' or not convertible to float
superstore_data = superstore_data[superstore_data['SALES'].str.lower() != 'unknown']

# Convert 'SALES' column to float
superstore_data['SALES'] = superstore_data['SALES'].replace('[\$,]', '', regex=True).astype(float)

# Drop rows with missing values in key columns
superstore_data.dropna(subset=['SALES', 'DISCOUNT', 'profit'], inplace=True)

# Group by discount level (rounded to nearest 0.1) and calculate average sales and profit
superstore_data['discount_level'] = superstore_data['DISCOUNT'].round(1)
discount_analysis = superstore_data.groupby('discount_level')[['SALES', 'profit']].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(discount_analysis['discount_level'], discount_analysis['profit'], marker='o', label='Average Profit')
plt.plot(discount_analysis['discount_level'], discount_analysis['SALES'], marker='x', label='Average Sales')
plt.title('Effect of Discount on Business Performance')
plt.xlabel('Discount Level')
plt.ylabel('Average Value')
plt.legend()
plt.grid(True)

# Set up the file path to save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "RAWP3V1"
save_path.mkdir(parents=True, exist_ok=True)

# Save the plot as a PNG file
save_file_path = save_path / "Trial4.png"
plt.tight_layout()  # Adjust layout to prevent overlap
plt.savefig(save_file_path)

# Show the plot
plt.show()
```
