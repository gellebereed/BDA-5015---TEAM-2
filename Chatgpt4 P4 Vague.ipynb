```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load data
superstore_data = pd.read_csv('Superstore.csv', encoding='latin1')

# Group by Category and sum the Sales
category_sales = superstore_data.groupby('Category')['Sales'].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(8, 6))
category_sales.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()

# Save the figure to the specified path
desktop = Path.home() / "Desktop"
save_path = desktop / "P4V1"
save_path.mkdir(parents=True, exist_ok=True)  # Ensure the folder exists
plt.savefig(save_path / "Trial2.png")

# Show the plot in notebook
plt.show()
```
