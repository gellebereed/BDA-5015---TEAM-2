```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

# Load the data
superstore_data = pd.read_csv('Superstore.csv', encoding='latin1')

# Set the path to save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "P3V2"
save_path.mkdir(parents=True, exist_ok=True)  # Ensure the folder exists

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=superstore_data, x='Discount', y='Profit', alpha=0.6)
plt.title('Relationship Between Discount and Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid(True)

# Save the plot to the desired location
plot_filename = save_path / "Trial1.png"
plt.savefig(plot_filename, dpi=300)
plt.close()

print(f"Plot saved to: {plot_filename}")
```
