```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the data
superstore_data = pd.read_csv('Superstore_RAW.csv', encoding='latin1')

# Clean the dataset: Remove rows with missing or invalid discount/profit values
cleaned_data = superstore_data[['DISCOUNT', 'profit']].copy()
cleaned_data = cleaned_data.dropna()
cleaned_data = cleaned_data[
    (cleaned_data['DISCOUNT'].apply(lambda x: isinstance(x, (int, float)))) & 
    (cleaned_data['profit'].apply(lambda x: isinstance(x, (int, float))))
]

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(cleaned_data['DISCOUNT'], cleaned_data['profit'], alpha=0.6)
plt.title('Scatter Plot of Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid(True)

# Set up the file path to save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "RAWP3V2"
save_path.mkdir(parents=True, exist_ok=True)

# Save the plot as a PNG file
save_file_path = save_path / "Trial1.png"
plt.tight_layout()  # Adjust layout to prevent overlap
plt.savefig(save_file_path)

# Show the plot
plt.show()
```
