```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the dataset
superstore_data = pd.read_csv('Superstore.csv', encoding='latin1')

# Convert 'Order Date' to datetime format
superstore_data['Order Date'] = pd.to_datetime(superstore_data['Order Date'])

# Filter data for the year 2016
data_2016 = superstore_data[superstore_data['Order Date'].dt.year == 2016]

# Group by Sub-Category and sum the Sales
sales_by_subcategory = data_2016.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)

# Create a bar chart for total sales by sub-category
plt.figure(figsize=(10, 6))
sales_by_subcategory.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Sub-Category in 2016', fontsize=14)
plt.xlabel('Sub-Category', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Set up the file path to save the plot
desktop = Path.home() / "Desktop"
save_path = desktop / "P4V2"
save_path.mkdir(parents=True, exist_ok=True)

# Save the plot as a PNG file
save_file_path = save_path / "Trial1.png"
plt.tight_layout()  # Adjust layout to prevent overlap
plt.savefig(save_file_path)

# Show the plot
plt.show()

# Print the save path for confirmation
print(f"Plot saved to {save_file_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the dataset
superstore_data = pd.read_csv('Superstore.csv', encoding='latin1')

# Convert 'Order Date' to datetime
superstore_data['Order Date'] = pd.to_datetime(superstore_data['Order Date'])

# Filter data for the year 2016
data_2016 = superstore_data[superstore_data['Order Date'].dt.year == 2016]

# Group by 'Sub-Category' and calculate total sales
sales_by_subcategory = data_2016.groupby('Sub-Category')['Sales'].sum().reset_index()

# Sort the data by total sales in descending order
sales_by_subcategory = sales_by_subcategory.sort_values('Sales', ascending=False)

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.barh(sales_by_subcategory['Sub-Category'], sales_by_subcategory['Sales'], color='skyblue')
plt.xlabel('Total Sales (USD)')
plt.title('Total Sales Across Product Sub-Categories in 2016')
plt.gca().invert_yaxis()  # Invert y-axis to display the highest sales at the top

# Define the save path for the image
desktop = Path.home() / "Desktop"
save_path = desktop / "P4V2"

# Ensure the save directory exists
save_path.mkdir(parents=True, exist_ok=True)

# Save the figure
save_path_image = save_path / "Trial2.png"
plt.savefig(save_path_image, bbox_inches='tight')

# Show the plot
plt.show()

print(f"Graph saved at: {save_path_image}")
```
